# petrol price Train/Test

```python
x=df[['Oil Barrel']]
y=df['Petrol Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
print(regression.score(x_test,y_test))
l_pred=regression.predict(x_test)
```

# SVR

```python
from sklearn.svm import SVR
regression_s=SVR()
regression_s.fit(x_train,y_train)
```

```python
print(regression_s.score(x_test,y_test))
s_pred=regression_s.predict(x_test)
```

# Decition tree

```python
from sklearn.tree import DecisionTreeRegressor
regression_d=DecisionTreeRegressor()
regression_d.fit(x_train,y_train)
```

```python
print(regression_d.score(x_test,y_test))
d_pred=regression_d.predict(x_test)
```

# random forest

```python
from sklearn.ensemble import RandomForestRegressor
regression_r=RandomForestRegressor()
regression_r.fit(x_train,y_train)
```

```python
print(regression_r.score(x_test,y_test))
r_pred=regression_r.predict(x_test)
```

# petrol price save model

```python
import pickle
filename='petrol_model.pkl'
pickle.dump(regression_s,open(filename,'wb'))
```
